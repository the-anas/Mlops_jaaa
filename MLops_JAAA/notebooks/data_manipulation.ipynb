```python
data1 = pd.read_csv('../data/raw/data_raw.csv')
```

```python
# adjust the settings so we can view all the columns
pd.set_option('display.max_columns', None)
# adjust the settings so we can view the first 2000 the rows
pd.set_option('display.max_rows', 2000)
```

```python
data1
```

```python
#remove all the columns that have more than 50% missing values
data1 = data1.loc[:, data1.isnull().mean() < .5]
```

```python
#remove all the rows that have missing values
data1 = data1.dropna()
```

```python
#remove all the columns that only contain 0 values
data1 = data1.loc[:, (data1 != 0).any(axis=0)]
```

```python
data1
```

```python
#drop the columns tradeImbalance|Ask 1-3 and Bid 1-3
data1 = data1.drop(['TradeImbalance|Ask1', 'TradeImbalance|Ask2', 'TradeImbalance|Ask3', 'TradeImbalance|Bid1', 'TradeImbalance|Bid2', 'TradeImbalance|Bid3'], axis=1)
```

```python
data1
```

```python
# group the data by ticker
grouped1 = data1.groupby('Ticker')
```

```python
#split the dataset into 5 datasets based on the ticker
for ticker, group in grouped1:
    group.to_csv('../data/processed/' + ticker + '.csv', index=False)
```
